```python
# This is mall customer segmentation using KMeans clustering algorithm
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Importing the dataset
df = pd.read_csv('../data/mall_customers.csv')

# Exploratory Data Analysis
print("The head of the data: ")
df.head()

print("\nThe tail of the data: ")
df.info()

print("\nThe shape of the data: ")
df.describe()

print("\nThe missing values of the data: ")
df.isnull().sum()
```

```python
df.drop(columns=['CustomerID', 'Genre'], inplace = True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

df_scaled = sc.fit_transform(df)

df_scaled
```

```python
pd.DataFrame(sc.inverse_transform(df_scaled), columns = df.columns)
```

```python
df_sc = pd.DataFrame(df_scaled, columns = df.columns)
df_sc.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.cluster import KMeans
KMeans()
```

```python
km = KMeans(n_clusters=5, random_state = 15)
```

```python
km.fit(df_sc)
```

```python
km.labels_
```

```python
km.cluster_centers_
```

```python
km.inertia_
```

```python
df['label'] = km.labels_
df.head()
```

```python
sns.scatterplot(data=df, x='Age', y='Annual_Income_(k$)', hue='label', palette=['red','blue','yellow','orange','black'] )
```

```python
df.describe()
```

```python
df[df['label']==1].describe()
```

```python
# Predicting the clusters
df['label'] = km.predict(df_sc)
df.head()
```

```python
# Visualising the clusters
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Age', y='Annual_Income_(k$)', hue='label', palette=['red','blue','yellow','orange','black'] )
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=300, c='black', marker='*', label='centroid')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.show()
```
