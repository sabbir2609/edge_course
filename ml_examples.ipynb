```python
import numpy as np
```

```python
def train_linear_regression_model(X, y):
    def calculate_weights(X, y):
        # Calculate the weights using the normal equation
        return np.linalg.inv(X.T @ X) @ X.T @ y
    
    def predict(X, weights):
        # make predictions using the calculated weights
        return X @ weights
    
    def calculate_mse(y_true, y_pred):
        # Calculate the mean squared error
        return np.mean((y_true - y_pred) ** 2)
    
    # Calculate the weights
    weights = calculate_weights(X, y)

    # Make predictions
    predictions = predict(X, weights)

    # Calculate the mean squared error
    mse = calculate_mse(y, predictions)

    return weights, predictions, mse
```

```python
# Sample usage
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

weights, predictions, mse = train_linear_regression_model(X, y)
print(f"Weights: {weights}")
print(f"MSE: {mse}")
```

```python
# Print sample data
print(f"Sample data: {y[:5].flatten()}")
print(f"Predictions: {predictions[:5].flatten()}")
```
